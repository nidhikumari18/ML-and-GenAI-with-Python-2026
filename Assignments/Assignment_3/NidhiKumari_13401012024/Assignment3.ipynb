**2. Create a function to calculate sum of first N natural numbers.**

```python
def sum(n):
    sum=0
    for i in range(1,n+1):
        sum+=i
    return sum
num=int(input("Enter the value of n:"))
result=sum(num)
print(result)
```

**3. Create a function to reverse a number.**

```python
def reverseNum(n):
  reverse_num=0
  while n!=0:
    digit=n%10
    reverse_num=reverse_num*10+digit
    n//=10
  return reverse_num
n=int(input("Enter the Number:"))
result=reverseNum(n)
print(result)


```

**4. Create a Function to count digits in a number.**

```python
def countDigits(n):
  count=0
  num=str(n)
  for i in num:
    count+=1
  return count
n=int(input("Enter the Number:"))
result=countDigits(n)
print(f"The Number of digits in {n} is {result}.")
```

**5. Create a Function to check Palindrome Numbers.**

```python
def palindrome(n):
  num=str(n)
  reversenum=0
  while n!=0:
    digit=n%10
    reversenum=reversenum*10+digit
    n//=10
  if str(reversenum)==num:
    return True
  else:
    return False
n=int(input("Enter the Number:"))
result=palindrome(n)
print(result)
```

**6. Create a Function to generate Fibonacci series.**

```python
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci(10)
```


**7. Calculator Using Functions that contains the following features:<br>
	-	User selects operation<br>
	-	Program performs calculation<br>
	-	Display result**


```python
def calc(a,b,c):
  result=0
  if c=='+':
    result=a+b
  elif c=='-':
    result=a-b
  elif c=='*':
    result=a*b
  else:
    result=a/b
  return result
a=int(input("Enter the first Number:"))
b=int(input("Enter the second Number:"))
c=input("Enter the operator(+,-,*,/):")
result=calc(a,b,c)
print(f"{a}{c}{b}={result}")
```

**8. Create a Text File and store Student Details.**

```python
def store_student_details():
    name = input("Enter Student Name: ")
    roll_no = input("Enter Roll Number: ")
    course = input("Enter Course: ")

    with open("student_details.txt", "w") as file:
        file.write("Student Details\n")
        file.write("-----------------\n")
        file.write(f"Name: {name}\n")
        file.write(f"Roll No: {roll_no}\n")
        file.write(f"Course: {course}\n")

    print("Student details stored successfully.")

store_student_details()
```

**10. Handle division by zero using exception handling.**

```python
def division(num1,num2):
  try:
    result = num1 / num2
    print("Result =", result)

  except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
a=int(input("Enter the first Number:"))
b=int(input("Enter the Second Number:"))
division(a,b)
```
